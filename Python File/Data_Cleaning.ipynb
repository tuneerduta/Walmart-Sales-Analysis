```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Walmart_new.csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
for i in df.columns:
    print(i,df[i].sort_values().unique(),'\n',sep='\n')
```

```python
df['Profit'] = df['Profit'].str.replace('Rs=','')
df['Profit'] = df['Profit'].str.replace('Rs','')
df['Product_ID'] = df['Product_ID'].str.replace('=','')
df['Region'] = df['Region'].replace('2West','West')
df['Region'] = df['Region'].replace('West#','West')
df['Ship_Date'] = df['Ship_Date'].str.replace('=','-')
df['Order_Date'] = df['Order_Date'].str.replace('=','-')
df['Customer_ID'] = df['Customer_ID'].str.replace('=','')
df['Order_ID'] = df['Order_ID'].str.replace('=','-')
```

```python
for i in df.columns:
    print(i,df[i].sort_values().unique(),'\n',sep='\n')
```

```python
df['Country'] = 'United States'
```

```python
df['Ship_Mode'] = df['Ship_Mode'].str.replace('StandardClass','Standard Class')
```

```python
df.drop('Row_ID',axis=1,inplace=True)
```

```python
df.dtypes
```

```python
date = ['Order_Date','Ship_Date']
integer = ['Postal Code','Quantity']
for i in date:
    df[i] = df[i].astype('datetime64[ns]')

df['Profit'] = df['Profit'].astype('float64')

for i in integer:
    df[i] = df[i].astype('int32')
```

```python
df.info()
```

```python
df.rename(columns = {'Postal Code':'Postal_Code','Sub=Category':'Sub_Category','Product Name':'Product_Name'}, inplace = True)
```

```python
df['Product_Name'] = df['Product_Name'].str.replace("="," ")
```

```python
result = []
my_list = list(df['Order_ID'])
for i in range(len(my_list)):
    res = my_list[i][2:7]
    rep = my_list[i].replace(str(res),'')
    result.append(rep)
result
df['Order_ID'] = result
```

```python
df.drop('Region',axis=1,inplace=True)
```

```python
df.to_csv('Walmart_Data.csv',index=False)
```
